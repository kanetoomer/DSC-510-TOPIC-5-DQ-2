```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# I'm generating synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X.flatten() + np.random.randn(100) * (1 + X.flatten()) 

# Fit linear regression model
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Showing plots for residuals vs. fitted values
plt.scatter(predictions, model.resid)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Fitted Values')
plt.axhline(0, color='red', linestyle='--')
plt.show()

# Weighted Least Squares to address heteroscedasticity
weights = 1 / (X[:, 1]**2)
model_wls = sm.WLS(y, X, weights=weights).fit()
predictions_wls = model_wls.predict(X)

# Showing plots for residuals vs. fitted values for WLS model
plt.scatter(predictions_wls, model_wls.resid)
plt.xlabel('Fitted values (WLS)')
plt.ylabel('Residuals (WLS)')
plt.title('Residuals vs. Fitted Values (WLS)')
plt.axhline(0, color='red', linestyle='--')
plt.show()
```
